```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#KOSPI

```python
df_KOSPI_200 = pd.read_csv(r"KOSPI 50 Historical Data.csv")
```

```python
df_KOSPI_200['Date'] = pd.to_datetime(df_KOSPI_200['Date'])
df_KOSPI_200 = df_KOSPI_200.sort_values(by='Date').reset_index(drop=True)
df_KOSPI_200.set_index('Date', inplace=True)
```

```python
df_KOSPI_200['Price'] = df_KOSPI_200['Price'].str.replace(',', '').str.replace(' ', '')
df_KOSPI_200['Price'] = pd.to_numeric(df_KOSPI_200['Price'], errors='coerce')
```

```python
prices = df_KOSPI_200['Price']
returns_KOSPI_200 = np.log(prices / prices.shift(1)).dropna()
returns_KOSPI_200 = returns_KOSPI_200.to_frame(name='Return')
```

#RESULT & VISUALIZATION

```python
returns_dict = {
    "ret_KOSPI (KRW)": returns_KOSPI_200,
    "ret_KOSPI_Exclude_COVID (KRW)": re_KOSPI_Norma,
    "ret_ASX (USD)": returns_ASX,
    "ret_ASX_Exclude_COVID (AUD)": re_ASX_Norm,
    "ret_FTSE_250 (GBP)":returns_FTSE_250,
    "ret_FTSE_250_Exclude_COVID (GBP)":re_FTSE_Norm,
    "ret_FTSE_100 (GBP)":returns_FTSE_100,
    "ret_FTSE_100_Exclude_COVID (GBP)":re_FTSE100_Norm
}

price_dict = {
    "ret_KOSPI (KRW)": df_KOSPI_200['Price'],
    "ret_KOSPI_Exclude_COVID (KRW)": df_KOSPI_Norm['Use'],
    "ret_ASX (USD)": df_ASX['Price'],
    "ret_ASX_Exclude_COVID (AUD)": df_ASX_Norm['Use'],
    "ret_FTSE_250 (GBP)":df_FTSE_250['Price'],
    "ret_FTSE_250_Exclude_COVID (GBP)":df_FTSE_Norm['Use'],
    "ret_FTSE_100 (GBP)":df_FTSE_100['Price'],
    "ret_FTSE_100_Exclude_COVID (GBP)":df_FTSE100_Norm['Use']
}
```

```python
max_abs_change_dict = {}

for name, series in returns_dict.items():
    if isinstance(series, pd.DataFrame):
        series = series.mean(axis=1)

    # หา abs เปลี่ยนแปลงแรงสุด
    abs_series = series.abs()
    max_idx = abs_series.idxmax()
    max_change = abs_series.loc[max_idx]

    # หา Pt และ Pt+1 จากราคาจริง (ต้องมี price_series[name] คู่กัน)
    try:
        price_series = price_dict[name]  # <- ต้องสร้าง price_dict ก่อนนะจ๊ะ
        pt = price_series.shift(1).loc[max_idx]
        pt1 = price_series.loc[max_idx]
    except:
        pt, pt1 = None, None

    max_abs_change_dict[name] = {
        "Max_Abs_Change": max_change,
        "Date": max_idx,
        "Pt": round(pt, 4) if pt is not None else None,
        "Pt+1": round(pt1, 4) if pt1 is not None else None
    }
```

```python
max_abs_change_dict
```

```python
tick_size_dict = {
    "KOSPI": 0.01,
    "ASX": 1,
    "FTSE":0.5
}

contract_size_dict = {
    "KOSPI": 250000,
    "ASX": 10,
    "FTSE":10
}
somc_dict = {
    "KOSPI": "10 Ticks * Tick Size * Contract Size",
    "ASX": " $0.50 per contract for certain equities",
    "FTSE":" SCRAF Ratio * SOM Rate"
}
```

```python
def extract_base_name(name):
    if "KOSPI" in name:
        return "KOSPI"
    elif "ASX" in name:
        return "ASX"
    elif "SET50" in name:
        return "SET50"
    elif "FTSE" in name:
        return "FTSE"
    else:
        return name
```

```python
summary_data = []

for name, detail in max_abs_change_dict.items():
    base = extract_base_name(name)

    tick = tick_size_dict.get(base, None)
    contract = contract_size_dict.get(base, None)
    somc = somc_dict.get(base, "Unknown")

    summary_data.append({
        "Instrument": name,
        "Tick_Size": tick,
        "Contract_Size": contract,
        "Max_Abs_Change": detail.get("Max_Abs_Change", None),
        "Date": detail.get("Date", None),
        "Pt": detail.get("Pt", None),
        "Pt+1": detail.get("Pt+1", None),
        "SOMC_Policy": somc
    })

# 📊 สร้าง DataFrame
df_summary = pd.DataFrame(summary_data).set_index("Instrument")
```
