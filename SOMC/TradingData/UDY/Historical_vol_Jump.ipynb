```python
file_path = r"SOMC Data SET Clear (1).xlsx"

df_vol = pd.read_excel(file_path, sheet_name="SET50 Volatility")

df_rf = pd.read_excel(file_path, sheet_name="Zero Yield (Rf)")

# ตรวจสอบข้อมูล
print(df_vol.head())
print(df_rf.head())
```

```python
df_vol['CALC_DATE'] = pd.to_datetime(df_vol['CALC_DATE'], format='%Y%m%d')
df_vol.set_index('CALC_DATE', inplace=True)

df_rf['BIZ_DATE'] = pd.to_datetime(df_rf['BIZ_DATE'], format='%Y%m%d')
df_rf.set_index('BIZ_DATE', inplace=True)

print(df_vol.head())
print(df_rf.head())
```

```python
df_rf_fix = df_rf[df_rf['TTM'] == 1.0][['YIELD']]
print(df_rf_fix.head())
```

```python
df_price = pd.read_csv(r"SET 50 Historical Data (7).csv")
```

```python
df_price['Date'] = pd.to_datetime(df_price['Date'])  
df_price.set_index('Date', inplace=True)             
```

```python
df_price
```

```python
df_price = df_price.sort_index(ascending=True)
df_price
```

```python
# remove commas from the 'Price' column and convert to numeric
df_price['Price'] = df_price['Price'].str.replace(',', '', regex=False)
df_price['Price'] = pd.to_numeric(df_price['Price'], errors='coerce')
```

```python
n_steps = 8
step_size = 25

strike_data = []

for date, row in df_price.iterrows():
    S0 = row['Price']
    
    if pd.isna(S0):  # ถ้า S0 เป็น NaN → ใส่ NaN ทั้งแถว
        strike_list = [np.nan] * (2 * n_steps + 1)
    else:
        S0_rounded = round(S0 / step_size) * step_size
        strike_range = np.arange(-n_steps, n_steps + 1) * step_size
        strike_list = S0_rounded + strike_range
    
    strike_data.append(strike_list)
strike_columns = [f"K{i - n_steps}" for i in range(2 * n_steps + 1)]
df_strike = pd.DataFrame(strike_data, index=df_price.index, columns=strike_columns)
```

```python
df_strike
```

```python
df_vol['VOL'] = df_vol['HIS_VLT'] / 100
df_rf_fix['RF'] = df_rf_fix['YIELD'] / 100


T = 0.25


common_dates = df_price.index.intersection(df_vol.index).intersection(df_rf_fix.index)
df_strike_filtered = df_strike.loc[common_dates]
df_price_filtered = df_price.loc[common_dates]
df_vol_filtered = df_vol.loc[common_dates]
df_rf_filtered = df_rf_fix.loc[common_dates]


df_nd1 = pd.DataFrame(index=common_dates, columns=df_strike.columns)


for date in common_dates:
    S = df_price_filtered.loc[date, 'Price']
    sigma = df_vol_filtered.loc[date, 'VOL']
    r = df_rf_filtered.loc[date, 'RF']
    K_list = df_strike_filtered.loc[date].values

    with np.errstate(divide='ignore', invalid='ignore'):
        d1 = (np.log(S / K_list) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
        nd1 = norm.cdf(d1)

    df_nd1.loc[date] = nd1
```

```python
df_nd1
```

```python
df_nd2 = pd.DataFrame(index=common_dates, columns=df_strike.columns)

for date in common_dates:
    S = df_price_filtered.loc[date, 'Price']
    sigma = df_vol_filtered.loc[date, 'VOL']
    r = df_rf_filtered.loc[date, 'RF']
    K_list = df_strike_filtered.loc[date].values

    with np.errstate(divide='ignore', invalid='ignore'):
        d1 = (np.log(S / K_list) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
        d2 = d1 - sigma * np.sqrt(T)
        nd2 = norm.cdf(d2)

    df_nd2.loc[date] = nd2
```

```python
df_nd2
```

```python
df_call = pd.DataFrame(index=common_dates, columns=df_strike.columns)
df_put = pd.DataFrame(index=common_dates, columns=df_strike.columns)

T =  0.25 

for date in common_dates:
    try:
        S = float(df_price_filtered.loc[date, 'Price'])
        K_list = df_strike_filtered.loc[date].astype(float).values
        r = float(df_rf_filtered.loc[date, 'RF'])
        sigma = float(df_vol_filtered.loc[date, 'VOL'])

        # --- d1 & d2 ---
        with np.errstate(divide='ignore', invalid='ignore'):
            d1 = (np.log(S / K_list) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
            d2 = d1 - sigma * np.sqrt(T)

        # --- N(d1) & N(d2) ---
        Nd1 = norm.cdf(d1)
        Nd2 = norm.cdf(d2)
        Nnd1 = norm.cdf(-d1)
        Nnd2 = norm.cdf(-d2)

        # --- Call / Put ---
        call = S * Nd1 - K_list * np.exp(-r * T) * Nd2
        put = K_list * np.exp(-r * T) * Nnd2 - S * Nnd1

        df_call.loc[date] = call
        df_put.loc[date] = put

    except Exception as e:
        print(f"Error on {date}: {e}")
        df_call.loc[date] = [np.nan] * len(df_strike.columns)
        df_put.loc[date] = [np.nan] * len(df_strike.columns)
```

```python
# สมมุติ shock เป็น +80%
shock_factor = 1.4

# สร้างคอลัมน์ใหม่
df_vol['VOL_Shock'] = df_vol['VOL'] * shock_factor
df_vol['VOL_Shock']
df_vol_filtered = df_vol.loc[common_dates]
```

```python
df_call_shock = pd.DataFrame(index=common_dates, columns=df_strike.columns)
df_put_shock = pd.DataFrame(index=common_dates, columns=df_strike.columns)

T = 0.25 

for date in common_dates:
    try:
        S = float(df_price_filtered.loc[date, 'Price'])
        K_list = df_strike_filtered.loc[date].astype(float).values
        r = float(df_rf_filtered.loc[date, 'RF'])
        sigma = float(df_vol_filtered.loc[date, 'VOL_Shock'])

        # --- d1 & d2 ---
        with np.errstate(divide='ignore', invalid='ignore'):
            d1 = (np.log(S / K_list) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
            d2 = d1 - sigma * np.sqrt(T)

        # --- N(d1) & N(d2) ---
        Nd1 = norm.cdf(d1)
        Nd2 = norm.cdf(d2)
        Nnd1 = norm.cdf(-d1)
        Nnd2 = norm.cdf(-d2)

        # --- Call / Put ---
        call = S * Nd1 - K_list * np.exp(-r * T) * Nd2
        put = K_list * np.exp(-r * T) * Nnd2 - S * Nnd1

        df_call_shock.loc[date] = call
        df_put_shock.loc[date] = put

    except Exception as e:
        print(f"Error on {date}: {e}")
        df_call_shock.loc[date] = [np.nan] * len(df_strike.columns)
        df_put_shock.loc[date] = [np.nan] * len(df_strike.columns)
```
