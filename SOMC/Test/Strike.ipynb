```python
USD = pd.read_excel(r"USd_BOT_2005-9Jun2025.xlsx")
```

```python
mx = round(USD['USD'].max()/0.25)*0.25
mx
```

```python
mn = round(USD['USD'].min()/0.25)*0.25
mn
```

```python
strike_range = np.arange(mn, mx+0.25, 0.25)
```

```python
strike_range
```

```python
strike_range = strike_range.flatten()
```

```python
strike_steps = 2
```

```python
strike_list = []
for i in range(len(USD)):
    Used = round(USD.loc[i, 'USD']/0.25) * 0.25
    USD.loc[i, 'Round'] = Used
    strike_range_compare = np.arange(-strike_steps, strike_steps + 1) * 0.25
    
    if i == 0:
        strike_list = Used + strike_range_compare
    else:
        strike_list = np.unique(np.concatenate([strike_list, Used + strike_range_compare]))
    
    for j in strike_range:
        USD.loc[i, j] = j in strike_list

```
