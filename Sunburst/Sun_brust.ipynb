```python
file_path = r"Before_split.xlsx"
X = pd.read_excel(file_path)
```

```python
X = X.drop(columns = "จำนวน (คน)")
```

```python
rows = []

for idx, row in X.iterrows():
    date = row["Date"]
    parti_id = row["PARTI_ID"]

    for col in X.columns:
        if col in ["Date", "PARTI_ID"]:
            continue

        code_part = col.split(" ")[0]
        code_levels = code_part.split(".")
        value = row[col]

        # ข้ามถ้า value ว่างเปล่าหรือไม่ใช่ตัวเลข
        if pd.isna(value):
            continue

        lvl_1_val = lvl_1.get(code_levels[0], None) if len(code_levels) >= 1 else None
        lvl_2_val = lvl_2.get(".".join(code_levels[:2]), None) if len(code_levels) >= 2 else None
        lvl_3_val = lvl_3.get(".".join(code_levels[:3]), None) if len(code_levels) >= 3 else None
        lvl_4_val = lvl_4.get(".".join(code_levels[:4]), None) if len(code_levels) == 4 else None

        # เก็บเฉพาะแถวที่ date/parti_id ยังอยู่
        if pd.notna(date) and pd.notna(parti_id):
            rows.append({
                "Date": date,
                "PARTI_ID": parti_id,
                "lvl_1": lvl_1_val,
                "lvl_2": lvl_2_val,
                "lvl_3": lvl_3_val,
                "lvl_4": lvl_4_val,
                "value": value
            })

used_df = pd.DataFrame(rows)
```

```python
used_df = used_df.fillna(method='ffill', axis=1)
used_df
```

```python
used_df.to_excel(r"Q_After_Split.xlsx", index=False)
```
