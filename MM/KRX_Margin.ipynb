```python
import pandas as pd
import re
import glob
import os

folder_path = '/Users/fulinq/Documents/SET/MarginModel/KRX'
os.chdir(folder_path)
```

```python
files = os.listdir(folder_path)

date_column_name = 'Effective Date'

list_of_dataframes = []
for file_name in files:

    if file_name.endswith('.xls') or file_name.endswith('.xlsx'):
        df = pd.read_excel(os.path.join(folder_path, file_name))
        
        # Remove extension for date parsing
        if file_name.endswith('.xlsx'):
            date_str = file_name[:-5]
        else:
            date_str = file_name[:-4]
        # Remove extra dots and spaces
        date_str = date_str.replace('.', '').strip()
        effective_date = pd.to_datetime(date_str)

        df[date_column_name] = effective_date
        list_of_dataframes.append(df)

combined_df = pd.concat(list_of_dataframes, ignore_index=True)
combined_df
```

```python
pivot_df = combined_df.pivot_table(index=date_column_name, columns='Product Name', values='Maintenance Margin Rate(%)' , aggfunc='first')
pivot_df = pivot_df.sort_index().fillna(method='ffill')
pivot_df
```
