```python
import pandas as pd
import os

folder_path = '/Users/fulinq/Documents/SET/MarginModel/Excel'
os.chdir(folder_path)
```

```python
df = pd.read_excel('Settlement_Price_Futures_2013-2025.xlsx', sheet_name='Settlement_Price')
print("\nOriginal Data (First 5 Rows):")
print(df.head())

df = df[['Date_format', 'Trim', 'Settlement_Price']]
df.rename(columns={'Date_format' : 'Date',
                   'Trim' : 'PROD_ULY_CODE'}, inplace=True)
df = df[df['Date'] >= '2015-01-01']
# df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
df['Settlement_Price'] = pd.to_numeric(df['Settlement_Price'], errors='coerce')
df.head(5)
```

```python
pivoted_df = df.pivot_table(index='Date', columns='PROD_ULY_CODE', values='Settlement_Price', aggfunc='first')
pivoted_df.reset_index(inplace=True)
pivoted_df
```

```python
with pd.ExcelWriter('2015_Pivot_Settlement_Price.xlsx') as writer:
    pivoted_df.to_excel(writer, index=False, sheet_name='Settlement_Price')
```
