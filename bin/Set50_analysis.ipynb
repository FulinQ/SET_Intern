```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('SET 50 Historical Data (3).csv')
data = data[['Date', 'Price']]
data.rename(columns={'Price': 'SET50'}, inplace=True)
data.tail(10)

set_data = pd.read_csv('USD_THB Historical Data (1).csv')
set_data = set_data[['Date', 'Price']]

drop_data = set_data.dropna(axis=1)
drop_data.tail(5)

print(data)
# print(drop_data.tail(5))
```

```python
# Convert 'Date' in the 'data' DataFrame (from CSV)
try:
    data['Date'] = pd.to_datetime(data['Date'])
    print("Converted 'Date' in 'data' (from CSV) to datetime.")
except Exception as e:
    print(f"Could not convert 'Date' in 'data' to datetime: {e}")
    # Handle error appropriately if conversion fails

# Convert 'Date' in the 'drop_data' DataFrame (from Excel)
try:
    # Check if 'Date' column exists after dropna
    if 'Date' in drop_data.columns:
        drop_data['Date'] = pd.to_datetime(drop_data['Date'])
        print("Converted 'Date' in 'drop_data' (from Excel) to datetime.")
    else:
        print("Warning: 'Date' column not found in 'drop_data'. Cannot merge on Date.")

except Exception as e:
    print(f"Could not convert 'Date' in 'drop_data' to datetime: {e}")
    # Handle error appropriately if conversion fails


# --- Merge the DataFrames on the 'Date' column ---
if 'Date' in data.columns and 'Date' in drop_data.columns:
    print("\nMerging 'data' and 'drop_data' on 'Date'...")
    # Use pd.merge and specify the column to merge on
    merged_df = pd.merge(data, drop_data, on='Date', how='inner') # 'inner' keeps only matching dates
```

```python
y_value = df['SET50']
x_value = df.drop(columns=['SET50', 'Date'], axis=1)
model = arch_model(y = y_value, x = x_value, mean= 'LS', vol = "GARCH", p = 1, q = 1, rescale = True)
res = model.fit(update_freq=5)
res.summary()
```

```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_sca = scaler.fit_transform(x_value)

lasso = LassoCV(cv = 5).fit(x_sca, y_value)

selected = lasso .coef_ != 0
x_las_used = x_value.loc[:, selected]
x_las_used
```

```python
model_2 = arch_model(y = y_value, x = x_las_used, mean = 'LS', vol = 'GARCH', p=1, q=1, rescale = True)
res_2 = model_2.fit()
res_2.summary()
```
